### Use Random Forest to parse only important attributes for the model out of all the columns. These are the one that influences decision the most in successful credit card applications. Then apply XGBoost with K-fold Cross Validation to build a model to predict succesful credit  card applications. Use the hyperparameter values as needed for formulation of the model.

```python
import pandas as pd
```

```python
df = pd.read_csv('credit_card_application.csv')
df.head()
```

```python
df.info()
```

```python
#Remove unnecessary columns from the dataset
df_features = df.drop(['ID','DAYS_BIRTH','DAYS_EMPLOYED','begin_month','target','FLAG_MOBIL','FLAG_WORK_PHONE',
                       'OCCUPATION_TYPE'], axis=1)
df_target = df['target']
```

```python
df_features
```

```python
df["target"].value_counts()
```

```python
#change to numbers
change_number = {"CODE_GENDER": {"M": 1, "F": 0},
                "FLAG_OWN_CAR": {"Y": 1, "N": 0},
                "FLAG_OWN_REALTY": {"Y": 1, "N": 0}}                        
```

```python
data = df_features.replace(change_number)
data.head()
```

```python
#get dummies for categorical variables
data_final = pd.get_dummies(data, columns=['NAME_INCOME_TYPE', 'NAME_EDUCATION_TYPE','NAME_FAMILY_STATUS','NAME_HOUSING_TYPE'])
```

```python
#display all
pd.set_option('display.max_rows', None)
data_final.head()
```

```python
data_final.shape
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

X= data_final  # Features
y= df['target'] # Labels

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train)

y_pred=rfc.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from matplotlib import pyplot
```

```python
# get importance
importance = rfc.feature_importances_
# summarize feature importance
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

### We can see the most important feature of the data is AMT_INCOME_TOTAL. That makes sense because credit card applications are evaluated by the income total of applicants

## Apply XGBoost with K-fold Cross Validation to build a model to predict succesful credit card applications

```python
import xgboost as xgb
from sklearn.metrics import mean_squared_error
```

```python
data_dmatrix = xgb.DMatrix(data=X,label=y)
```

```python
params = {"objective":"reg:squarederror",'colsample_bytree': 0.3,'learning_rate': 0.1,
                'max_depth': 5, 'alpha': 10}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3,
                    num_boost_round=50,early_stopping_rounds=10,
                    metrics="rmse", as_pandas=True, seed=123)
```
